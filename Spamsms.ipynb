```python
import pandas as pd

file_path = ("C:/Users/85job/Downloads/spam (1).csv")

data_df = pd.read_csv(file_path, encoding='latin-1')

print(data_df)
```

```python
print(data_df.dtypes)
```

```python
pd.set_option('display.max_rows',None)
print(data_df)
```

```python
#data preparation
import numpy as np
data_df =  data_df.drop("Unnamed: 2" , axis=1)
data_df =  data_df.drop("Unnamed: 3" , axis=1)
data_df =  data_df.drop("Unnamed: 4" , axis=1)
```

```python
pd.set_option('display.max_rows',None)
print(data_df)
```

```python
data_df.isnull()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode 'ham' as 0 and 'spam' as 1
label_encoder = LabelEncoder()
data_df['v1'] = label_encoder.fit_transform(data_df['v1'])
```

```python
print(data_df)
```

```python
print(data_df['v2'].isnull().sum())  
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data
X_tfidf = tfidf_vectorizer.fit_transform(data_df['v2'])
```

```python
print(X_tfidf.shape) 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, data_df['v1'], test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Predict on the test set
y_pred = model.predict(X_test)

# Generate the confusion matrix
print(confusion_matrix(y_test, y_pred))

# Generate a detailed classification report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_tfidf, data_df['v1'], cv=5)  # 5-fold cross-validation
print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
```

```python
import numpy as np

# Get feature names from the vectorizer
feature_names = tfidf_vectorizer.get_feature_names_out()

# Get the log probabilities of each feature per class
class_log_prob = model.feature_log_prob_


top_n = 10
for i, class_prob in enumerate(class_log_prob):
    sorted_indices = np.argsort(class_prob)[-top_n:]  
    top_features = [feature_names[idx] for idx in sorted_indices]
    print(f"Top {top_n} features for class {i}:", top_features)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

param_grid = {'C': [0.01, 0.1, 1, 10, 100]}  
grid_search = GridSearchCV(LogisticRegression(), param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best parameters:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)
```
